```python
import pandas as pd
import numpy as np

movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
movies.head(3)
```

```python
credits.head(3)
```

```python
pd.set_option("display.max_column",24)
```

```python
data = pd.merge(left = movies,right = credits,on = 'title')
data.head(2)
```

```python
data["release_date"] = pd.to_datetime(data["release_date"])
```

```python
data.head(2)
```

```python
data["Year"] = data["release_date"].dt.year
```

```python
# Necessary features which will be helpfull to create grouped features based on which recommendation will be created:
cols = ['title','genres','id','keywords','overview','Year',"cast","crew"]
data = data[cols]
data.head(3)
```

```python
data = data.dropna()
```

```python
data.head(2)
```

```python
data["Year"] = data["Year"].astype("int")
```

```python
data["Year"] = data["Year"].astype("str")
```

```python
# Preprocessing the final dataset:
data.duplicated().sum()
```

```python
data.iloc[18]["genres"]
```

```python
import ast
def convert(obj):
    genres = []
    arr = ast.literal_eval(obj)
    for i in arr:
        genres.append(i["name"])
    return genres

data["genres"] = data["genres"].apply(convert)
```

```python
def convert_cast_crew(obj):
    lst = []
    c = 0
    arr = ast.literal_eval(obj)
    for i in arr:
        if c != 3:
            lst.append(i["name"])
            c += 1
        else:
            break
    return lst
```

```python
# Only obtain directors
def convert_crew(obj):
    lst = []
    c = 0
    arr = ast.literal_eval(obj)
    for i in arr:
        if i["job"] == "Director":
            lst.append(i["name"])
    return lst
```

```python
data.head(2)
```

```python
data["keywords"] = data["keywords"].apply(convert)
```

```python
data["cast"] = data["cast"].apply(convert_cast_crew)
```

```python
data["crew"] = data["crew"].apply(convert_crew)
```

```python
data["overview"] = data["overview"].apply(lambda x : x.split())
```

```python
data.head(2)
```

```python
data["Year"] = data["Year"].apply(lambda x:x.split())
```

```python
data["cast"] = data["cast"].apply(lambda x:[i.replace(" ","") for i in x])
data["overview"] = data["overview"].apply(lambda x:[i.replace(" ","") for i in x])
data["crew"] = data["crew"].apply(lambda x:[i.replace(" ","") for i in x])
data["keywords"] = data["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
data["genres"] = data["genres"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
data.head(2)
```

```python
data["tags"] = data["overview"] + data["genres"] + data["keywords"] + data["cast"] + data["crew"] + data["Year"]
```

```python
final = data[["id","title","tags"]]
final.head()
```

```python
final["tags"] = final["tags"].apply(lambda x:" ".join(x))
```

```python
final.head()
```

```python
final.to_csv("final_movies.csv",index = False)
```

```python
final["tags"] = final["tags"].str.lower()
```

```python
# Text vectorization on the tags:
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english',max_features = 5000)
```

```python
# Matrix of movies vs features:
vectors = cv.fit_transform(final["tags"]).toarray()
```

```python
vectors[2]
```

```python
# Which words are frequently occuring in the tags:
cv.get_feature_names()
```

```python
import nltk
```

```python
# Applying stemming operation on the top of tags column:
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemming(text):
    lst = []  
    for i in text.split():
        lst.append(ps.stem(i))
    res = " ".join(lst)
    return res

final["tags"] = final["tags"].apply(stemming)
```

```python
vectors
```

```python
# Computing cosine similarity:
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
# Similarity between first movie and all 4806 movies:
sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x : x[1])
```

```python
def get_names_of_movies(index):
    return final.iloc[index]["title"]
```

```python
# Recommend movies:
def Recommend(movie):
    movie_index = final[final["title"] == movie].index[0]
    distances = similarity[movie_index]
    ans = sorted(list(enumerate(distances)),reverse = True,key = lambda x : x[1])
 
    name = []
    for i in ans[:11]:
        name.append(get_names_of_movies(i[0]))
    name.pop(0)
    return name

# Top 10 Recommended Movies:
Recommend("The Amazing Spider-Man")
```
